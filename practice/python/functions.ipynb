```python
d1 = pd.DataFrame(
    {
        "X": ["X1","X2","X3","X4"],
        "Y": ["Y1","Y2","Y3","Y4"],
        "Z": ["Z1","Z2","Z3","Z4"],
    },
    index = [1,2,3,4]
)
d2 = pd.DataFrame(
    {
        "X": ["X5","X6","X7","X8"],
        "Y": ["Y5","Y6","Y7","Y8"],
        "Z": ["Z5","Z6","Z7","Z8"],
    },
    index = [5,6,7,8]
)
d3 = pd.DataFrame(
    {
        "X": ["X9","X10","X11","X12"],
        "Y": ["Y9","Y10","Y11","Y12"],
        "Z": ["Z9","Z10","Z11","Z12"],
    },
    index = [9,10,11,12]
)
```

```python
res = pd.concat( [d1, d2, d3])
res
```

```python
frames = [d1, d2, d3]
res = pd.concat( frames, keys=['a', 'b', 'c'])
res
```

```python
d4 = pd.DataFrame(
    {
        "X" : ["X2", "X5", "X6"],
        "Y" : ["Y2", "Y5", "Y6"],
        "W" : ["W2", "W5", "W6"]
    },
    index = [2, 5, 6]
)
d4
```

- An outer join will take a union of the datasets. Axis = 1 denotes the union of DataFrame with respect to column labels, and axis = 0 denotes the union of DataFrame with respect to row labels (index).

```python
pd.concat([d1, d4], axis = 1, join = "outer") # axis 1 means union of DF wrt col labels
```

```python
pd.concat([d1, d4], axis = 0, join = "outer") # axis 0 means union of DF wrt row labels
```

- An inner join will take an intersection of the datasets. Axis = 1 denotes the intersection of DataFrame with respect to column labels, and axis = 0 denotes the intersection of DataFrame with respect to row labels (index).

```python
pd.concat([d1, d4], axis = 1, join = "inner")
```

```python
pd.concat([d1, d4], axis = 0, join = "inner")
```

**PLOT**

```python
pdf = pd.DataFrame( np.random.randn( 15, 4), columns=['a', 'b', 'c', 'd'])
pdf.plot(figsize=(15, 4))
```

```python
# creating subplots using pyplot

figure, axes = plt.subplots(2,2)

# setting height and width of plot
figure.set_figheight(12)
figure.set_figwidth(15)

pdf.plot.bar(ax= axes[0][0])

pdf.plot.bar(stacked=True, ax= axes[0][1])

pdf.plot.barh(ax= axes[1][0])

pdf.plot.barh(stacked= True, ax= axes[1][1])
```

```python
p_df = pd.DataFrame(
    {
        "a": np.random.randn(1000) + 1,
        "b": np.random.randn(1000),
        "c": np.random.randn(1000) - 1,
    },
    columns=["a", "b", "c"],
)
p_df
```

```python
figure, axes = plt.subplots(2,2)
figure.set_figheight(12)
figure.set_figwidth(15)

# simple histogram 
p_df.plot.hist(alpha=0.3, ax = axes[0][0])

# stacked histogram
p_df.plot.hist(stacked = True, ax = axes[0][1])

# horizontal histogram
p_df.plot.hist(orientation = "horizontal", ax=axes[1][0], cumulative = True, color = "red", alpha = 0.4)

# histogram using .diff() method for a column
p_df["a"].diff().hist()
```
