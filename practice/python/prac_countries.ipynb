```python
import urllib.request
import pandas as pd
import numpy as np
```

```python
from urllib.request import urlretrieve

ur = urlretrieve('https://gist.githubusercontent.com/aakashns/28b2e504b3350afd9bdb157893f9725c/raw/994b65665757f4f8887db1c85986a897abb23d84/countries.csv', 
            'countries.csv')
```

```python
print( ur, ur[0])
```

```python
df = pd.read_csv( ur[0])
df.head()
```

```python
# Q1: How many countries does the dataframe contain?

locdf = df['location']
# print( locdf )
print( locdf.is_unique) # since the values are unique the locdf has all the unique values
```

```python
#  Q2: Retrieve a list of continents from the dataframe?

contdf = df['continent']
# print( contdf.head() )
cdf = pd.Series( contdf.unique() )
cdf
```

```python
# Q3.1: What is the total population of all the countries listed in this dataset?

popdf = df['population']
totalpop = popdf.sum()
totalpop
```

```python
# Q3.2: (Optional) What is the overall life expectancy across in the world?

lifedf = (df['life_expectancy'] * df['population']).sum() / totalpop
lifedf
```

```python
# Q4: Create a dataframe containing 10 countries with the highest population.

high10df = df.sort_values(by=['population']).head( 10)
high10df
```

```python
# Q5.1: Add a new column in countries_df to record the overall GDP per country (product of population & per capita GDP).

df['gdp'] = df['population'] * df['gdp_per_capita']
df.head()
```

```python
# Q5.2: (Optional) Create a dataframe containing 10 countries with 
# the lowest GDP per capita, among the counties with population greater than 100 million.

lowestdf = df[df.population > 100000000].sort_values( by=['gdp'], ascending= True).head(10)
lowestdf
```

```python
# Q6: Create a data frame that counts the number countries in each continent?

countrydf = df.groupby(['continent'])['location'].count()
countrydf
```

```python
# Q7: Create a data frame showing the total population of each continent.

totalcontpop = df.groupby(['continent'])['population'].sum()
totalcontpop
```

```python
# **covid data**


ur2 = urlretrieve('https://gist.githubusercontent.com/aakashns/b2a968a6cfd9fbbb0ff3d6bd0f26262b/raw/b115ed1dfa17f10fc88bf966236cd4d9032f1df8/covid-countries-data.csv', 
            'covid-countries-data.csv')
df2 = pd.read_csv( ur2[0])
df2.tail()
```

```python
# Q8: Count the number of countries for which the total_tests data is missing.

# Hint: Use the .isna method.

missingdf = df2['total_tests'].isna().sum()
missingdf
# df2.shape
```

```python
# Q10: Add columns tests_per_million, cases_per_million and deaths_per_million into combined_df.

combineddf = df.merge( df2, on = 'location' )
combineddf.head()
```

```python

combineddf['total_cases_perM'] = combineddf['total_cases']*1e6 / combineddf['population']
combineddf['total_deaths_perM'] = combineddf['total_deaths']*1e6 / combineddf['population']
combineddf['total_tests_perM'] = combineddf['total_tests']*1e6 / combineddf['population']

combineddf.head()
```

```python
# Q11: Create a dataframe with 10 countires that have highest number of tests_per_million people.

hightpm = combineddf.sort_values(by=['total_tests_perM'], ascending = False).head(10)
hightpm
```

```python
# Q12: Create a dataframe with 10 countires that have highest number of positive cases per million people.

highcpm = combineddf.sort_values(by=['total_cases_perM'], ascending = False).head(10)
highcpm
```

```python
# Q13: Create a dataframe with 10 countires that have highest number of deaths cases per million people?

highdpm = combineddf.sort_values(by=['total_deaths_perM'], ascending = False).head(10)
highdpm
```

```python
# Q14: Count number of countries that feature in both the lists of "highest number of tests per million" 
# and "highest number of cases per million".

list1 = set(highdpm['location'])
list2 = set(highcpm['location'])
res1 = list1.intersection(list2)
print( len( res1))
```

```python
# Q15: Count number of countries that feature in both the lists "20 countries with lowest GDP per capita" 
# and "20 countries with the lowest number of hospital beds per thousand population". 

# Only consider countries with a population higher than 10 million while creating the list.

lowestgdp = combineddf.sort_values(by=['gdp'], ascending = True).head(20)
lowesthos = combineddf.sort_values(by=['hospital_beds_per_thousand'], ascending = True).head(20)
gt10 = combineddf[ combineddf.population > 1000000]
# print( lowestgdp['location'], lowesthos['location'], gt10['location'])

li3 = set ( lowestgdp['location'] ).intersection( lowesthos['location'])
li4 = set ( li3.intersection( gt10['location']) )
print( li4 )
```
