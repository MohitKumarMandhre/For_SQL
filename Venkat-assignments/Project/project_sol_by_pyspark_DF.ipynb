```
import re

def convert_into_min(time):
    if not isinstance(time,str):
        return 0
    else:
        words = time.split()
        total_mins = 0
        
        if len(words) > 2:
            total_mins = int(re.sub('\D','',words[0])) * 60 + int(re.sub('\D','',words[2]))
        else:
            total_mins = int(re.sub('\D','',words[0]))

    return total_mins

convertoMIN = udf(convert_into_min,IntegerType())
```

```
mdf4 = mdf.withColumn('totalMinutes', convertoMIN(col('Running_time')) )
mdf4.show()
```

```
res = mdf4[['Movie_name', 'Year_of_release', 'totalMinutes']].orderBy(col('totalMinutes'), ascending=False).head(1)
print( res[0][0], res[0][1], res[0][2] )
```
